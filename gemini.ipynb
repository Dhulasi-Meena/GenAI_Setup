# Requirements

```python
import google.generativeai as genai
import os
```

```python
api=os.getenv('GOOGLE_API_KEY')
```

### List of all the model versions that support content generation

```python
for m in genai.list_models():
    if 'generateContent' in m.supported_generation_methods:
        print(m.name)
```

### Generate a content from Gemini model

```python
genai.configure(api_key=api)
```

```python
model=genai.GenerativeModel('gemini-2.5-flash-lite')
```

```python
prompt='''Assume you are a sports expert. Prepare a summary on MS Dhoni's carrer. Include all
 the key achievemnets and statistics. Generate output in bullet points'''
```

```python
response=model.generate_content(prompt)
```

```python
print(response.text)
```
